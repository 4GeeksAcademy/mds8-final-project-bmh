```python

for cicle in range(len(data_all)):
    url_demog_ = data_all[cicle]
    time.sleep(10) 




    response = requests.get(url_demog_)  

    if response.status_code == 200:  
        with open(file_names[cicle] + ".xpt", 'wb') as file:  
            file.write(response.content)
            
            try:
                df, meta = pyreadstat.read_xport(file_names[cicle] + ".xpt", encoding='latin1')
                df.to_csv(file_names[cicle]+ ".csv", index = False)
                

                print("File downloaded successfully and converted to csv.")

                os.remove(file_names[cicle] + ".xpt")
                time.sleep(20)
            except Exception as e:  
                print(f"Error processing '{url_demog}': {e}") 
    else:  
        print(f"Failed to download file from '{url_demog}'. Status code: {response.status_code}") 
```

```python
code_mean = ["https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/DR1IFF_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/DR2IFF_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/BAX_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/BPXO_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/BMX_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/HDL_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/TCHOL_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/HSCRP_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/INS_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/GLU_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/FERTIN_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/GHB_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/VID_L.htm",
"https://wwwn.cdc.gov/Nchs/Data/Nhanes/Public/2021/DataFiles/CBC_L.htm"
]
results = {}
for code in code_mean:  
    print(f"Starting processing for URL: {code}")  
    time.sleep(20)
    try:
        response = requests.get(code)
        html_content = response.text

        soup = BeautifulSoup(html_content, "html.parser")

        title_text = soup.title.text.strip()

        title_data = {}

        print(title_text)
        print(soup.title)

        soup.title = {}
        lis = soup.find_all("li")
        for li in lis:
            a_tag = li.find("a")
            if a_tag:
                code_ = a_tag.text.strip().split("\n")
                if code_:
                    code_string = code_[0]
                    parts = code_string.split(" - ",1)
                    if len(parts)==2:
                        code, description = parts
                        title_data[code] = description


        results[title_text] = title_data
        print(soup.title)
    except requests.exceptions.RequestException as e:  
        print(f"An error occurred during the request: {e} for URL: {code}")  
    except Exception as e:  
        print(f"An error occurred: {e}") 
          
    
```
